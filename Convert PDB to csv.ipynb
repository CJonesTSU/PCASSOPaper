```python
import sys
import os
```

```python
class PDBEntry:
    def __init__(self, idx, name, chain):
        self.indexNum = idx
        self.name = name
        self.chain = chain
```

```python
class helix:
    def __init__(self, idx, chain, start, end,length):
        self.id = idx
        self.chain = chain
        self.length = length
        self.location_list = []
        for x in range(int(start),int(end)+1):
            self.location_list.append(x)
            #print(location_list)
        
```

```python
class Sheet:
    def __init__ (self, strand, sheetID, startChain, start, endChain, end):
        self.strand = strand
        self.sheetID = sheetID
        self.startChain = startChain
        self.start = start
        self.endChain = endChain
        self.end = end
        self.location_list = []
        for x in range(int(start),int(end) + 1):
            self.location_list.append(x)
#             print(location_list)
```

```python
class carbonAlpha:
    def __init__ (self, chain_id, seqNum, AA_type):
        self.chain_id = chain_id
        self.seqNum = seqNum
        self.AA_type = AA_type
```

```python
notebook_path = os.path.abspath("Convert PDB to csv.ipynb")

PDBList = os.path.join(os.path.dirname(notebook_path), "DataSet/Test_Set_300.csv")
PDBs = []
with open(PDBList) as file:
    for item in file:
        item = item.rstrip('\n')
        tmp = item.split(',')
        obj = PDBEntry(tmp[0], tmp[1],tmp[2])
        PDBs.append(obj)
print(len(PDBs))
```

```python
for pdb in PDBs:
    helicies = []
    sheets = []
    seq_nums = []
    helix_seq_nums = ''
    sheet_seq_nums = ''
    print(pdb.name)
    read_path = os.path.join(os.path.dirname(notebook_path),"PDBFiles",pdb.name)
    write_path = read_path
    read_path = read_path + "/" + pdb.name + ".pdb"
    write_path += '/' + pdb.name + '-pdb' +'.csv'
    out_file = open(write_path, 'w')
    in_file = open(read_path, 'r')
    lines = in_file.readlines()
    for line in lines:
        if(line[:5] == "HELIX"):
            idx = line[8:11].strip()
            chain = line[19:20].strip()
            start = line[21:25].strip()
            end = line[33:37].strip()
            length = line[71:76].strip()

            helicies.append(helix(idx, chain, start, end,length))
        elif(line[:5] == "SHEET"): 
            strand = int(line[8:10])
            sheetID = line[12:14].strip()
            startChain = line[21].strip()

            initSeqNum = line[22:26].strip()
            endChain = line[32].strip()
            endSeqNum = line[33:37].strip()
            sheets.append(Sheet(strand , sheetID, startChain, initSeqNum, endChain, endSeqNum))
        elif(line[:4] == 'ATOM'):
            if line[13:15] == 'CA':
                thisChain = line[21]
                thisSeqNum = int(line[22:26].strip())
                this_AA_type = line[17:20].strip()
                seq_nums.append(carbonAlpha(thisChain, thisSeqNum, this_AA_type))
    
    helix_seq_nums = []
    for h in helicies:
        for l in h.location_list:
            #print(f'l is: {l}')
            helix_seq_nums.append(l)
    helix_seq_nums.sort()

    sheet_seq_nums = []
    for s in sheets:
        for l in s.location_list:
            sheet_seq_nums.append(l)
    sheet_seq_nums.sort()

    out_file.write(f'{pdb.name}\n')
    for ca in seq_nums:
        ss_type = 'C'
        if ca.seqNum in helix_seq_nums:
            ss_type = 'H'
        elif ca.seqNum in sheet_seq_nums:
            ss_type = 'E'
        if ca.chain_id == pdb.chain:    
            out_file.write(f'{ca.seqNum},{ca.chain_id},{ca.AA_type},{ss_type}\n')

    out_file.close()
    in_file.close()
```
